```python
import re

def remove_units(name):
    name = re.sub(r'\d+[gpKmLlin].*$', '', name)
    return name

producto_chunk['NombreProductoCat'] = producto_chunk['NombreProducto'].apply(remove_units)

cat_names = list(producto_chunk['NombreProductoCat'].unique())
categories = {}

for product in cat_names:
    categories[product] = cat_names.index(product)
    
producto_chunk['Category'] = producto_chunk['NombreProductoCat'].map(categories)
producto_chunk
```
